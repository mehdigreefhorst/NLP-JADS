# The idea of using LLama

BERTopic is made to be very flexible, in the paper it is stated that BERTopic will beccome better as time goes on because the document encoder can be swapped by any other embedding model. The larger the model is, the more accurate the embedding, (Kaplan et al. 2020). Therefore, we thought that by implementing one of the largest large language models to embed our documents, the BERTopic model would become even better. THe model is huge, it takes a long time to download. The model chosen is llama3-8b, it is a large language model with 8 billion parameters

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
from bertopic import BERTopic
import torch

pipe = pipeline("feature-extraction", model="WillHeld/DiVA-llama-3-v0-8b", trust_remote_code=True, device="cuda")
topic_model = BERTopic(embedding_model=pipe)
```
